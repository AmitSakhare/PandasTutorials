```python
#DataFrame - is two dimensional array. Pandas DataFrame consists of three components is Data,rows and Column

#impoort pandas
import pandas as pd
```

```python
# creating list of strings
lst = ['Amit','Poonam','Ankita','Yash','Vijay']
df = pd.DataFrame(lst)
df
```

```python
#Dealing with Rows and column
#Cerating Dictionary for Student Data
stu_data = {
    'Name' : ['Jay','Ram','Santosh','Anju','Riya'],
    'Age':[25,28,20,33,26],
    'Address':['Aurangabad','Mumbai','Pune','Allahbad','Delhi'],
    'Course' :['BBA','BE','BA','MCA','MBA']
}
stu=pd.DataFrame(stu_data)
stu
```

```python
#Accessing Name and Age Column
#single column
print(stu['Age'])
#Multiple column
print(stu[['Name','Age']])
```

```python
#loc() and iloc() function - is a indexed based selection method that we have to pass specific values of rows and columns

stu.iloc[0] # accesing first rows
```

```python
stu.iloc[1:3] #accessing 1st and 2nd row
```

```python
# Add new column 
stu['Gender']=['Male','Male','Male','Female','Female']
stu
```

```python
# remove or drop column
stu.drop('Gender',axis=1,inplace=True)
stu
```

```python
# Select rows where Age is greater than 30
print(stu[stu['Age']>30])
```

```python
# import diabetes data
db_data = pd.read_csv("Data/diabetes.csv")
db_data.head(5)
```

```python
# Select age & BMI coloumn
age=db_data[['Age','BMI']].head()
age
```

```python
#sorting by coloumn
sort=db_data.sort_values(by='BMI')
sort
```

```python
print(db_data.isnull())
```

```python
#rename() function - we can rename various features names or multiple features at the same time
db_data.rename(columns={"Age":"age"})
```

```python
# drop() function - deleting specific column or multiple columns at the same time
db_data.drop( labels=['Outcome'], #  labels :index or name of the column you want drop
             axis=1 ,# axix: select column its bydefault value is 0
             inplace=True # inplace : if false than return a copt.otherwise retun none
             )
db_data
```

```python
import numpy as np
df1=pd.DataFrame([[
    np.nan, 2, np.nan, 0], 
     [3, 4, np.nan, 1], 
     [np.nan, np.nan, np.nan, 5], 
     [np.nan, 3, np.nan, 4]], 
     columns=list("ABCD"))
df1
```
